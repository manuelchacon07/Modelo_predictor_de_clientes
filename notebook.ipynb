## Preprocesamiento

```python
# Importamos las librerias necesarias para este apartado

import pandas as pd
```

```python
# Cargamos el csv y lo mostramos por pantalla

leads = pd.read_csv('Leads.csv')

leads
```

```python
# Listamos el tipo de columna

leads.dtypes
```

Vemos que cada columna tiene el tipo correcto, primero vamos a comprobar si tenemos valores duplicados en Lead Number y Prospect ID

```python
# Comprobamos si tenemos valores duplicados en Lead Number y Prospect ID

print(sum(leads.duplicated(subset = 'Prospect ID')) == 0)
print(sum(leads.duplicated(subset = 'Lead Number')) == 0)
```

Vemos que ambas columnas tienen valores duplicados, por lo que lo eliminaremos ya que no tiene sentido tener clientes duplicados

```python
# Eliminamos las columnas Lead Number y Prospect ID

leads = leads.drop(['Prospect ID', 'Lead Number'], axis = 1)
```

Vamos a comprobar el contenido de cada columna

```python
# Mostramos por pantalla los contenidos de cada una de las columnas

for i in leads.columns:
    print(leads[i].value_counts(dropna = False))
    print('-----------------------------------')
```

Vemos que tenemos muchas columnas que parecen que no son relevantes para nuestra prediccion, vamos a analizar sus contenidos..

```python
# Revisamos las columnas Do Not Call, Do Not Email, Search, Magazine, Newspaper Article, Receive More Updates About Our Courses, X Education Forums, Newspaper, Digital Advertisement,
# Through Recommendations, Receive More Updates About Our Courses, Update me on Supply Chain Content, Get updates on DM Content, I agree to pay the amount through cheque y
# A free copy of Mastering The Interview para ver si son relevantes

print(leads['Do Not Call'].value_counts(dropna = False))
print(leads['Search'].value_counts(dropna = False))
print(leads['Magazine'].value_counts(dropna = False))
print(leads['Newspaper Article'].value_counts(dropna = False))
print(leads['Receive More Updates About Our Courses'].value_counts(dropna = False))
print(leads['X Education Forums'].value_counts(dropna = False))
print(leads['Newspaper'].value_counts(dropna = False))
print(leads['Digital Advertisement'].value_counts(dropna = False))
print(leads['Through Recommendations'].value_counts(dropna = False))
print(leads['Receive More Updates About Our Courses'].value_counts(dropna = False))
print(leads['Update me on Supply Chain Content'].value_counts(dropna = False))
print(leads['Get updates on DM Content'].value_counts(dropna = False))
print(leads['I agree to pay the amount through cheque'].value_counts(dropna = False))
print(leads['Do Not Email'].value_counts(dropna = False))
print(leads['A free copy of Mastering The Interview'].value_counts(dropna = False))
```

Vemos que casi todo es "no", esto nos crearia ruido, provocando falsos negativos/positivos en nuestro modelo, por lo que los eliminaremos

```python
# Eliminamos las columnas que podrian crear ruido

leads = leads.drop(['Do Not Call', 'Do Not Email', 'Search', 'Magazine', 'Newspaper Article', 'Receive More Updates About Our Courses', 'X Education Forums', 'Newspaper',
                     'Digital Advertisement', 'Through Recommendations', 'Receive More Updates About Our Courses', 'Update me on Supply Chain Content',
                     'Get updates on DM Content', 'I agree to pay the amount through cheque', 'A free copy of Mastering The Interview'], axis = 1)
```

Vamos a comprobar las columnas que tenemos..

```python
# Comprobamos las columnas

leads.info()
```

Vemos que hay muchas columnas que contienen la palabra "Select", eso significa que no contiene nada ya que el cliente no ha elegido ninguna opcion por lo que transformaremos "Select" en valores null

```python
# comprobamos los valores null

import numpy as np
leads = leads.replace('Select', np.nan)
```

Vamos a comprobar las columas que contienen valores null

```python
# Comprobamos las columas que contienen valores null

columnas_null = leads.isna().sum().to_frame(name='nº de null').sort_values('nº de null', ascending=False)
columnas_null
```

Vemos que hay columnas que tienen mas de 3000 null, el tratarlas resultaria costoso ya que cada cliente es diferente por lo que procedemos a eliminarlas

```python
# Eliminamos las columnas con mas de 3000 nulls

for i in leads.columns:
    if leads[i].isna().sum()>3000:
        leads.drop(i, axis=1, inplace=True)
```

Despues de la limpieza, vamos a tratar las columas que usaremos

```python
# Comprobamos el contenido de las columnas

for i in leads.columns:
    print(leads[i].value_counts(dropna = False))
    print('-----------------------------------')
```

```python
# cambiamos los valores null por el valor mas indicado
 
leads['Lead Source'].fillna('Google', inplace=True)
leads['TotalVisits'].fillna(0.0, inplace=True)
leads['Country'].fillna('unknown', inplace=True)
leads['Page Views Per Visit'].fillna(0.0, inplace=True)
leads['Last Activity'].fillna('Not Specified', inplace=True)
leads['What is your current occupation'].fillna('Unemployed', inplace=True)
leads['What matters most to you in choosing a course'].fillna('Not Specified', inplace=True)
```

```python
# Agrupamos datos similares y valores que aparecen pocas veces

leads['Lead Source'] = leads['Lead Source'].replace('google','Google')
leads['Lead Source'] = leads['Lead Source'].replace('Facebook','Social Media')
leads['Lead Source'] = leads['Lead Source'].replace(['bing','Click2call','Press_Release',
                                                     'youtubechannel','welearnblog_Home',
                                                     'WeLearn','blog','Pay per Click Ads',
                                                    'testone','NC_EDM','Live Chat'] ,'Others')
```

```python
# Quitamos los espacios de todo el dataset

leads.columns = leads.columns.str.lower().str.replace(' ', '_')

string_columns = list(leads.dtypes[leads.dtypes == 'object'].index)
for col in string_columns:
    leads[col] = leads[col].str.lower().str.replace(' ', '_')
```

## Entrenamiento

```python
# Importamos las librerias necesarias para este apartado

from sklearn.compose import make_column_selector as selector
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
```

```python
# Guardamos en una variable la columna objetivo

target = leads.converted
```

```python
# eliminamos la columna objetivo
target = "converted"
target_colum = leads[target]

leads.drop(target, axis=1, inplace=True)
```

```python
# Separamos las columnas por categoricas y numericas

categorical_columns_selector = selector(dtype_include=object)
categorical_columns = categorical_columns_selector(leads)
print(categorical_columns)

numerical_columns_selector = selector(dtype_exclude=object)
numerical_columns = numerical_columns_selector(leads)
print(numerical_columns)
```

```python
# Asignamos un preprocesador para cada tipo de columna

categorical_preprocessor = OneHotEncoder(handle_unknown="ignore")
numerical_preprocessor = StandardScaler()
```

```python
# Creamos un pipeline preprocesador

preprocessor = ColumnTransformer([
    ('one-hot-encoder', categorical_preprocessor, categorical_columns),
    ('standard_scaler', numerical_preprocessor, numerical_columns)])
```

```python
# Cremamos el modelo

model = make_pipeline(preprocessor, LogisticRegression(max_iter=500))
```

```python
# Dividimos los datos

data_train, data_test, target_train, target_test = train_test_split(leads, target_colum)
```

```python
# Entrenamos el modelo

model.fit(data_train, target_train)
```

```python
# Mostramos por pantalla la efectividad del modelo

model.score(data_test, target_test)
```
